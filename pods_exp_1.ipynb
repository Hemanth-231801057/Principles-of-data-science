```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
df=pd.DataFrame(data)
```

```python
print(df.head())
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].ffill()
df.head()
```

```python
df['Age'] = df['Age'].bfill()
df.head()
```

```python
df['Cabin'] = df['Cabin'].fillna("unknown")
df.head()
```

```python
df["Sex_encoded"]=np.where(df["Sex"]=="male",1,0)
```

```python
df=df.drop_duplicates()
```

```python
mean = np.mean(df["Fare"])
std_dev = np.std(df["Fare"])
scaled_fare = (df["Fare"] - mean) / std_dev
print(scaled_fare)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
features1 = ['Pclass', 'Sex_encoded', 'Age', 'SibSp']
sns.pairplot(df[features1])
plt.show()
```

```python
features2 = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
correlation_matrix = df[features2].corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```
